# Splitting the data into test set and train set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
x_train=m.fit_transform(x_train)
x_test=m.transform(x_test)
```

# Using different methods for analysing the model

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
l=LogisticRegression(random_state=0)
l.fit(x_train,y_train)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
dt.fit(x_train,y_train)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion='entropy',n_estimators=10,n_jobs=-1)
rf.fit(x_train,y_train)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knn.fit(x_train,y_train)
```

```python
#SVM
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
gn=GaussianNB()
gn.fit(x_train,y_train)
```

```python
def accuracy(x_train,y_train):
    print("Logistic Regression Accuracy: ",l.score(x_train,y_train))
    print("K Nearest Neighbors Accuracy: ",knn.score(x_train,y_train))
    print("Naive-Bayes accuracy: ",gn.score(x_train,y_train))
    print("Support Vector Machines Accuracy: ",sv.score(x_train,y_train))
    print("Decision Tree Classifier Accuracy: ",dt.score(x_train,y_train))
    print("Random Forest Classifier Accuracy: ",rf.score(x_train,y_train))
accuracy(x_train,y_train)        
```

```python
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix
import sklearn.metrics as metrics
```

```python
def evaluate_preds(y_test,y_pred):
    accuracy=accuracy_score(y_test,y_pred)
    r2=r2_score(y_test,y_pred)
    print(f"Accuracy:{accuracy*100:.2F}%")
    print(f"r2_score:{r2:.2F}")
    
```

```python
def model(y_test,x_test):
    print("Evaluation of Logistic Regression: ")
    evaluate_preds(y_test,l.predict(x_test))
    print("Evaluation of K-nearest-neighbours: ")
    evaluate_preds(y_test,knn.predict(x_test))
    print("Evaluation of Naive-Bayes: ")
    evaluate_preds(y_test,gn.predict(x_test))
    print("Evaluation of SVM:")
    evaluate_preds(y_test,sv.predict(x_test))
    print("Evaluation of Decision Tree: ")
    evaluate_preds(y_test,dt.predict(x_test))
    print("Evaluation of Random Forest: ")
    evaluate_preds(y_test,rf.predict(x_test))
```

```python
model(y_test,x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
def conf(y_test,x_test):
    print("Logistic Regression: ")
    print(confusion_matrix(y_test,l.predict(x_test)))
    print("K-nearest-neighbours: ")
    print(confusion_matrix(y_test,knn.predict(x_test)))
    print("Naive-Bayes: ")
    print(confusion_matrix(y_test,gn.predict(x_test)))
    print("Support Vector Machines: ")
    print(confusion_matrix(y_test,sv.predict(x_test)))
    print("Decision Tree: ")
    print(confusion_matrix(y_test,dt.predict(x_test)))
    print("Random Forest: ")
    print(confusion_matrix(y_test,rf.predict(x_test)))
```

```python
conf(y_test,x_test)
```

# SVM is choosen as the best model

```python
cm=confusion_matrix(y_test,sv.predict(x_test))
def conf(mat):
    fig,ax=plt.subplots(figsize=(3,3))
    ax=sns.heatmap(mat,annot=True,cbar=False,)
    plt.xlabel("True label")
    plt.ylabel("Predicted label")
    #fix the broken annotations
    bottom,top=ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top-0.5)
conf(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,sv.predict(x_test)))
```

```python
import sklearn.metrics as metrics
fpr,tpr,threshold=metrics.roc_curve(y_test,sv.predict(x_test))
roc_auc=metrics.auc(fpr,tpr)
```
